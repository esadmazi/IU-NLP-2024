## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import defaultdict

class BPETokenizer:
    """
    Subword tokenizer using Byte-Pair Encoding.

    Attributes:
        vocab_size (int): Desired size of the subword vocabulary.
        word_frequencies (defaultdict): Word frequencies in the corpus.
        splits (dict): Split representations of words.
        merges (dict): Merged subwords.
    """

    def __init__(self, vocab_size: int) -> None:
        """
        Initialize BPETokenizer.

        Args:
            vocab_size (int): Desired size of the subword vocabulary.
        """
        self.vocab_size = vocab_size
        self.word_frequencies = defaultdict(int)  # Dictionary to store word frequencies
        self.splits = {}  # Dictionary to store split representations of words
        self.merges = {}  # Dictionary to store merged subwords

    def train(self, corpus: List[str]) -> None:
        """
        Train BPETokenizer on given corpus.

        Args:
            corpus (List[str]): Corpus to train the tokenizer on.
        """
        self._analyze_corpus(corpus)
        self._build_subword_vocabulary()

    def _analyze_corpus(self, corpus: List[str]) -> None:
        """
        Analyze corpus to compute word frequencies.

        Args:
            corpus (List[str]): Corpus to analyze.
        """
        for text in corpus:
            words = str(text).split()
            for word in words:
                self.word_frequencies[word] += 1

    def _build_subword_vocabulary(self) -> None:
        """
        Build subword vocabulary.

        Constructs subword vocabulary by iteratively merging
        the most frequent character pairs until desired size.
        """
        # Extract characters from corpus
        alphabet = set(''.join(self.word_frequencies.keys()))  
        # Initialize vocabulary with special token and sorted characters
        vocab = ["</w>"] + sorted(alphabet)  
         # Initialize split representations of words
        self.splits = {word: list(word) for word in self.word_frequencies.keys()} 
        
        while len(vocab) < self.vocab_size:
            # Compute frequencies of character pairs
            pair_frequencies = self._compute_pair_frequencies()  
            # Find the most frequent character pair
            best_pair = max(pair_frequencies, key=pair_frequencies.get) 
            # Merge the most frequent character pair
            self.splits = self._merge_best_pair(*best_pair) 
            # Store merged subwords
            self.merges[best_pair] = best_pair[0] + best_pair[1]  
            # Update vocabulary with merged subword
            vocab.append(best_pair[0] + best_pair[1])  

    def _compute_pair_frequencies(self) -> defaultdict:
        """
        Compute frequencies of character pairs.

        Returns:
            defaultdict: Frequencies of character pairs.
        """
        # Dictionary to store frequencies of character pairs
        pair_frequencies = defaultdict(int)  
        for word, freq in self.word_frequencies.items():
             # Split representation of the word
            split = self.splits[word] 
            if len(split) == 1:
                continue
            for i in range(len(split) - 1):
                 # Character pair
                pair = (split[i], split[i + 1]) 
                pair_frequencies[pair] += freq
        return pair_frequencies

    def _merge_best_pair(self, a, b) -> dict:
        """
        Merge most frequent character pair.

        Args:
            a (str): First character of pair.
            b (str): Second character of pair.

        Returns:
            dict: Updated splits after merging pair.
        """
        for word in self.word_frequencies:
            # Split representation of the word
            split = self.splits[word]  
            if len(split) == 1:
                continue
            i = 0
            while i < len(split) - 1:
                if split[i] == a and split[i + 1] == b:
                    split = split[:i] + [a + b] + split[i + 2 :]
                else:
                    i += 1
            # Update split representations with merged subword
            self.splits[word] = split  
        return self.splits

    def tokenize(self, text: str) -> List[str]:
        """
        Tokenize input text using trained tokenizer.

        Args:
            text (str): Input text to tokenize.

        Returns:
            List[str]: List of subword tokens.
        """
        # List to store subword tokens
        tokens = [] 
        for word in text.split():
            # Split representation of the word
            splits_text = [list(word)]  
            for pair, merge in self.merges.items():
                for idx, split in enumerate(splits_text):
                    i = 0
                    while i < len(split) - 1:
                        if split[i] == pair[0] and split[i + 1] == pair[1]:
                            split = split[:i] + [merge] + split[i + 2 :]
                        else:
                            i += 1
                    splits_text[idx] = split
            tokens.extend(sum(splits_text, []))
        return tokens
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import os

data_folder = "Assignment1_data"

tweet_list = []

for file_num in range(1, 6):
    file_path = os.path.join(data_folder, f"file{file_num}")
    with open(file_path, 'r', encoding='utf-8') as file:
        tweets = file.readlines()
        tweet_list.extend(tweets)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Vocab sizes
vocab_sizes = [100, 250, 500, 750, 1000]

# Dictionary to store average tokenized tweet lengths for each vocab size
avg_lengths = {}

for vocab_size in vocab_sizes:
    
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(tweet_list)
    tokenized_tweets = [tokenizer.tokenize(tweet) for tweet in tweet_list]
    avg_length = np.mean([len(tokens) for tokens in tokenized_tweets])
    
    if vocab_size in avg_lengths:
        avg_lengths[vocab_size] += avg_length
    else:
        avg_lengths[vocab_size] = avg_length
        
# Plot the dependency of average tokenized tweet length on vocab size
plt.plot(list(avg_lengths.keys()), list(avg_lengths.values()), marker='o')
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Tokenized Tweet Length')
plt.title('Dependency of Average Tokenized Tweet Length on Vocabulary Size(Original Dataset)')
plt.grid(True)
plt.show()        
```

## Toughts on the result:
Generally, a balance needs to be struck between a smaller vocabulary size (which may lead to loss of information) and a larger vocabulary size (which may lead to increased computational complexity).

For our task, since our dataset is consist of only 220 tweets, 1000 as vocab_size is the most effective.
